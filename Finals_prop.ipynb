```python
def merge_unemp_diab(unemp, diab):
    # unemp 데이터를 melt하여 연도별로 정리
    unemp_melted = unemp.melt(id_vars=['행정구역별'], var_name='연도', value_name='실업률')
    unemp_melted['연도'] = unemp_melted['연도'].astype(int)
    
    # dial 데이터를 melt하여 연도별로 정리
    diab_melted = diab.melt(id_vars=['행정구역별'], var_name='연도', value_name='조율')
    diab_melted['연도'] = diab_melted['연도'].astype(int)
    
    # 두 데이터프레임을 행정구역별과 연도별로 병합
    merged_df = pd.merge(unemp_melted, diab_melted, on=['행정구역별', '연도'], how='outer')
    
    return merged_df

# 병합 실행
merged_df = merge_unemp_diab(unemp, diab)

merged_df
```

```python
# NaN 및 '-' 값을 포함한 행 제거
merged_df.replace('-', pd.NA, inplace=True)  # '-' 값을 NaN으로 대체
merged_df.dropna(inplace=True)  # NaN 값을 포함한 행 제거

merged_df
```

```python
from shapely.geometry import Polygon
from shapely.ops import unary_union

df = sig

# --구 만 있는 행 제거
df = df[~df['SIG_KOR_NM'].str.match(r'^[가-힣]+구$')]
df['SIG_KOR_NM'].unique()

# --시 --구 형태를 --시로 변환하고 그룹화
df['City'] = df['SIG_KOR_NM'].apply(lambda x: x.split()[0] if '시' in x else None)
df['City'].unique()

# 그룹화된 데이터 프레임 생성
city_groups = df.dropna(subset=['City']).groupby('City')

# 평균 geometry 계산 함수
def average_geometry(polygons):
    return unary_union(polygons).centroid

# 새로운 데이터 프레임 생성
new_data = {
    'SIG_KOR_NM': [],
    'geometry': []
}

# 각 그룹별로 처리
for name, group in city_groups:
    new_data['SIG_KOR_NM'].append(name)
    new_data['geometry'].append(average_geometry(group['geometry']))

# 나머지 --시 --구 아닌 행들을 추가
other_rows = df[df['City'].isna()]

new_df = pd.concat([pd.DataFrame(new_data), other_rows.drop(columns=['City'])], ignore_index=True)
new_df = new_df[['SIG_KOR_NM', 'geometry']]
new_df
```

```python
# 열 이름 변경
sido = sido.rename(columns={'CTP_KOR_NM': '행정구역별'})
new_df = new_df.rename(columns={'SIG_KOR_NM': '행정구역별'})

# 두 데이터 프레임을 위아래로 붙이기
result_df = pd.concat([sido, new_df], ignore_index=True)
result_df
```

```python
# 좌표 참조 시스템 설정 (예: EPSG:5179)
crs = {'init': 'epsg:5179'}
geo = gpd.GeoDataFrame(result_df, crs=crs, geometry='geometry')

# geo['center_point'] = geo['geometry'].geometry.centroid
# geo['geometry'] = geo['geometry'].to_crs(epsg=4326)
# geo['center_point'] = geo['center_point'].to_crs(epsg=4326)
# geo['경도'] = geo['center_point'].map(lambda x: x.xy[0][0])
# geo['위도'] = geo['center_point'].map(lambda x: x.xy[1][0])

# 중심점 계산
geo['center_point'] = geo['geometry'].centroid

# 좌표 참조 시스템을 EPSG:4326으로 변환
geo = geo.to_crs(epsg=4326)
geo['center_point'] = geo['center_point'].to_crs(epsg=4326)

# 경도와 위도 추출
geo['경도'] = geo['center_point'].map(lambda x: x.x)
geo['위도'] = geo['center_point'].map(lambda x: x.y)


geo
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from pysal.lib import weights
from pysal.model import spreg

# Queen’s contiguity 방식으로 공간 가중치 행렬 생성
w = weights.Queen.from_dataframe(geo_data, use_index=True)

# 가중치 행렬 확인
print(w.full()[0])  # 첫 번째 행정구역의 인접성 행렬 출력
```

## SAR 모델 결과 해석
SAR (Spatial Autoregressive Model) 모델의 결과는 다음과 같습니다. 이 모델은 공간적 자기 상관성을 고려하여 회귀분석을 수행한 결과입니다.

### 주요 지표 해석

- Data set: 분석에 사용된 데이터셋의 이름이 Health Data Analysis로 명시되어 있습니다.
- Weights matrix: 공간 가중치 행렬의 이름이 Queen contiguity weights로 명시되어 있습니다.
- Dependent Variable (종속 변수): 당뇨병 유병률입니다.
- Number of Observations (관찰치 수): 1839개의 관찰치가 사용되었습니다.
- Mean dependent var (종속 변수의 평균): 종속 변수(당뇨병 유병률)의 평균 값은 11.9896입니다.
- S.D. dependent var (종속 변수의 표준편차): 종속 변수의 표준편차는 2.7881입니다.
- Pseudo R-squared (의사 R-제곱값): 모델의 설명력을 나타내며, 0.2356입니다. 이는 독립 변수가 종속 변수의 변동성을 약 23.56% 설명함을 의미합니다.
- Spatial Pseudo R-squared (공간적 의사 R-제곱값): 공간적 요인을 포함한 모델의 설명력으로, 0.2307입니다. 이는 공간적 상관성을 고려한 모델이 종속 변수의 변동성을 약 23.07% 설명함을 의미합니다.

### 변수별 결과 해석

#### CONSTANT (상수항):

- Coefficient (계수): 13.58841
- Std.Error (표준오차): 0.16992
- z-Statistic: 79.97092
- Probability (유의확률): 0.00000
상수항의 계수는 13.58841입니다. 이는 다른 변수가 0일 때 당뇨병 유병률의 예상 값을 나타냅니다. 상수항은 매우 유의미한 결과를 보여줍니다 (p-value = 0.00000).

#### 실업률 (독립 변수):

- Coefficient (계수): -0.90281
- Std.Error (표준오차): 0.04784
- z-Statistic: -18.86945

실업률의 계수는 -0.90281입니다. 이는 실업률이 1% 증가할 때 당뇨병 유병률이 약 0.903% 감소함을 의미합니다. 이 결과는 매우 유의미합니다 (p-value = 0.00000).

### 공간 회귀 모델에서 로버스트 표준 오류 사용

```python
import spreg

# 종속 변수와 독립 변수 설정
y = data_gdf['당뇨병 유병률'].values
X = data_gdf[['실업률']].values
X = sm.add_constant(X)

# 공간 가중 행렬 생성 (Queen contiguity)
w = Queen.from_dataframe(data_gdf, use_index=True)
w_name = "Queen contiguity weights"

# SAR 모델 적용 (로버스트 표준 오류 포함)
sar_model_rob = spreg.GM_Lag(y, X, w=w, name_y="당뇨병 유병률", 
                         name_x=["const", "실업률"], name_w=w_name,
                         name_ds="Health Data Analysis", robust='white')

# SAR 모델 요약 출력
print(sar_model_rob.summary)
```
